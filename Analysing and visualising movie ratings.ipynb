```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
dataset = pd.read_csv("RottenTomatoesMovies.csv")
dataset
```

```python
dataset.isnull().sum()
```

```python

dataset.dropna(inplace = True)
```

```python
dataset
```

```python
duplicates = dataset.duplicated().sum()
```

```python
dataset.drop_duplicates(inplace = True)
```

```python
dataset.tail()
```

```python
mean_rating= dataset['audience_rating'].mean()
mean_rating
```

```python
median_rating= dataset['audience_rating'].median()
median_rating
```

```python
mode_rating= dataset['audience_rating'].mode()[0]
mode_rating
```

```python
#Data Vissualisation
#Histogram of ratings

plt.figure(figsize = (10, 6))
sns.histplot(dataset['audience_rating'], bins = 20, kde= True)
plt.xlabel("Rotten Tomatoes Ratings")
plt.ylabel('Number of Movies')
plt.title('Distribution of movie ratings')
plt.show()
```

```python
plt.figure(figsize = (8,6))
sns.boxplot(dataset['audience_rating'])
plt.ylabel('ROTTEN TOMATOES RATINGS')
plt.title('Box Plot of Movie Ratings')
plt.show()
```

```python
#top rated movies 
top_rated_movies = dataset[dataset['audience_rating'] > 95].sort_values(by = 'audience_rating', ascending =False)
top_rated_movies
```

```python
#Average rating by genre

dataset['genre'] = dataset['genre'].astype(str)
dataset_exploded = dataset.explode('genre')

genre_ratings = dataset_exploded.groupby('genre')['audience_rating'].mean().sort_values(ascending =False)

plt.figure(figsize=(12, 9))
sns.barplot(x=genre_ratings.index[:10], y=genre_ratings.index[:10])
plt.xticks(rotation =90)
plt.xlabel("Genre")
plt.ylabel("Average rating")
plt.title("Top 10 Genere by Average")
plt.show()

```
